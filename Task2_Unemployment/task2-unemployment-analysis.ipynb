```python
df = pd.read_csv("/kaggle/input/datasets/gokulrajkmv/unemployment-in-india/Unemployment in India.csv")
df.head()
```

```python
df.columns
```

```python
df[' Date'] = pd.to_datetime(df[' Date'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.lineplot(x=' Date', y=' Estimated Unemployment Rate (%)', data=df)
plt.xticks(rotation=90)
plt.title("Unemployment Rate Over Time")
plt.show()
```

```python
covid_data = df[df[' Date'].dt.year >= 2020]

plt.figure(figsize=(12,6))
sns.lineplot(x=' Date', y=' Estimated Unemployment Rate (%)', data=covid_data)
plt.xticks(rotation=90)
plt.title("Unemployment Rate During Covid-19 Period")
plt.show()
```

```python
## Conclusion

The unemployment rate increased significantly during the Covid-19 period.
The visualization clearly shows fluctuations and peak unemployment levels in 2020.
This analysis highlights the impact of the pandemic on employment trends.
```
